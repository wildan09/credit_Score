# Home Credit Indonesia

### Task Background:

Many people find it difficult to get a loan due to poor or non-existent credit history. Home Credit strives to expand financial inclusion for the unbanked by providing a positive and secure lending experience. To ensure these underserved communities have a positive lending experience. Home Credit uses a variety of alternative data to predict customers' repayment ability. By doing so, Home Credit will ensure that customers who are able to repay are not rejected and loans are provided with a principal, maturity and repayment calendar that will empower customers to be successful.

### Goal:

1. Identify the characteristics of potential customers who will have difficult loan repayment.
2. Predict the repayment ability of customers

### Action :

1. Perform data cleansing, and visualization for business insights.
2. Build models with machine learning algorithms.
3. Predict customer repayment ability based on application test data with the best machine learning model.
4. Provide recommendations to the company to increase the success of customers in applying for loans.

It can be seen that the distribution of the numerical features mentioned above is skewed, so the missing values in those features will be accounted for by the median.

```python
# impute missing values with median because the data is skewed for numerical features
# impute missing values with mode for categorical features

category_columns = housing_df.select_dtypes(include=['object']).columns.tolist()
integer_columns = housing_df.select_dtypes(include=['int64','float64']).columns.tolist()

for column in housing_df:
    if housing_df[column].isnull().any():
        if(column in category_columns):
            housing_df[column]=housing_df[column].fillna(housing_df[column].mode()[0])
        else:
            housing_df[column]=housing_df[column].fillna(housing_df[column].median())
```

```python
# after imputation
print('Missing values status:', housing_df.isnull().values.any())
print('\nThe number of missing values for each columns (after imputation): \n')
nvc = pd.DataFrame(housing_df.isnull().sum().sort_values(), columns=['Total Null Values'])
nvc['Percentage'] = (nvc['Total Null Values']/housing_df.shape[0])*100
nvc.sort_values(by=['Percentage'], ascending=False).reset_index()
```
